## 함수 작성
- 입력된 url에서 자료 읽고 해당 상품에 연계된 관심상품의 가격 출력
- 현재 시각 추가

```python
import time

def 가격산출(url, 관심상품):
	response = requests.get(url)
	환전자료 = json.loads(response.text)['data']['rates']
	가격 = [str(time.asctime())]
	for 상품 in 관심상품:
		가격 += [1.0/float(환전자료[상품])]
	return 가격
```

```python
결과 = 가격산출(url, 관심상품)
print(결과)
```

```python
from apscheduler.schedulers.background import BackgroundScheduler
```

```python
from apscheduler.triggers.interval import IntervalTrigger
```

```python
sched = BackgroundScheduler()

def 가격산출():
    url = 'https://api.coinbase.com/v2/exchange-rates?currency=KRW'
    관심상품 = ['USD', 'EUR', 'JPY', 'CNY', 'BTC', 'ETH', 'DOGE']
    저장파일 = open('원화-환율.csv', 'a', encoding = 'utf-8')
    response = requests.get(url)
    환전자료 = json.loads(response.txt)['data']['rates']
    가격 = str(time.asctime())
    for 상품 in 관심상품:
        가격 += ', '+ str(1.0/float(환전자료[상품]))
    print(가격)
    저장파일.write(가격+'\n')
    저장파일.close()
```

```python
sched.start()
sched.add_job(가격산출, IntervalTrigger(seconds = 600), id = 'test')
```

```python
sched.shutdown(가격산출)
```
